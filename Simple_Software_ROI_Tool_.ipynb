```python
import ipywidgets as widgets
from IPython.display import display, clear_output

# Function to calculate ROI
def calculate_roi(cost, revenue_increase, time_saved, employee_hourly_rate):
    additional_revenue = revenue_increase
    cost_savings = time_saved * employee_hourly_rate
    total_benefit = additional_revenue + cost_savings
    roi = ((total_benefit - cost) / cost) * 100
    return roi, total_benefit

```

```python
# Create input widgets
cost_input = widgets.FloatText(description='Tool Cost ($):', value=0)
revenue_input = widgets.FloatText(description='Revenue Increase ($):', value=0)
time_saved_input = widgets.FloatText(description='Time Saved (hours):', value=0)
rate_input = widgets.FloatText(description='Hourly Rate ($):', value=0)
calculate_button = widgets.Button(description='Calculate ROI', button_style='success')
reset_button = widgets.Button(description='Reset Form', button_style='warning')
output = widgets.Output()
```

```python
# Define button click event for calculate
@output.capture(clear_output=True)
def on_calculate_clicked(b):
    try:
        cost = cost_input.value
        revenue_increase = revenue_input.value
        time_saved = time_saved_input.value
        hourly_rate = rate_input.value

        # Perform calculation
        roi, total_benefit = calculate_roi(cost, revenue_increase, time_saved, hourly_rate)

        # Display results
        print(f"Total Benefit: ${total_benefit:,.2f}")
        print(f"ROI: {roi:.2f}%")
    except Exception as e:
        print(f"Error: {str(e)}")
```

```python
# Define button click event for reset
@output.capture(clear_output=True)
def on_reset_clicked(b):
    cost_input.value = 0
    revenue_input.value = 0
    time_saved_input.value = 0
    rate_input.value = 0
    print("Form has been reset.")

# Attach events to buttons
calculate_button.on_click(on_calculate_clicked)
reset_button.on_click(on_reset_clicked)

# Display widgets and output
display(widgets.VBox([cost_input, revenue_input, time_saved_input, rate_input, widgets.HBox([calculate_button, reset_button]), output]))
```
